```python
data = data.rename(columns={'name': 'Province'})
data
```

```python
# set to datetime values
data['datetime'] = pd.to_datetime(data['datetime'])
```

```python
data['year'] = data['datetime'].dt.year
data
```

```python
data['max_temp_year'] = data.groupby(['Province', 'year'])['tempmax'].transform(max)
data['min_temp_year'] = data.groupby(['Province', 'year'])['tempmin'].transform(min)
data['avg_temp_year'] = data.groupby(['Province', 'year'])['temp'].transform(np.mean)
data['avg_humid_year'] = data.groupby(['Province', 'year'])['humidity'].transform(np.mean)
data
```

```python
data = data.rename(columns={'year': 'InspectionYear'})
```

```python
data.to_csv('temperature.csv', index=False)
```
